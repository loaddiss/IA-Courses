# LE MACHINE LEARNING


## 1. Des Limites des systèmes experts

Les systèmes experts (SE) sont des programmes informatiques conçus pour simuler les compétences et le raisonnement d'un expert humain dans un domaine spécifique. Bien que ces systèmes aient prouvé leur efficacité dans de nombreuses applications (comme le diagnostic médical, l'analyse financière, etc.), ils présentent plusieurs limites :

#### 1.1. **Dépendance aux règles explicites**
   - Les systèmes experts fonctionnent selon des règles précises codifiées par des experts humains. Ces règles sont généralement en nombre limité et ne peuvent pas toujours capturer toute la complexité d'un problème réel. Par exemple, un SE pourrait ne pas être capable de traiter des situations inédites ou des cas extrêmes qui ne correspondent pas aux règles existantes.
   
#### 1.2. **Manque de flexibilité**
   - Les systèmes experts sont rigides dans leur fonctionnement. Si une nouvelle règle ou un nouvel aspect du domaine d'expertise doit être pris en compte, il est nécessaire de modifier le système de manière explicite, ce qui est coûteux et prend du temps. De plus, ils ne sont pas capables de s’adapter à de nouvelles informations de manière autonome.
   
#### 1.3. **Difficulté à gérer les connaissances imprécises ou incomplètes**
   - Les SE sont basés sur des connaissances structurées et formelles, ce qui les rend peu efficaces face à l’incertitude, aux imprécisions ou aux informations manquantes. Par exemple, dans des situations où il manque des données essentielles ou lorsqu’il existe des ambiguïtés, un SE peut donner des résultats erronés ou incapables de prendre une décision.
   
#### 1.4. **Coût et complexité de la mise à jour des connaissances**
   - La mise à jour des bases de connaissances dans un système expert peut être très coûteuse et exige une expertise considérable. La maintenance du système devient ainsi un défi, particulièrement lorsque l'expertise dans le domaine évolue rapidement. Cela empêche le système de rester au fait des dernières découvertes ou innovations.

#### 1.5. **Manque d'intuition et de raisonnement créatif**
   - Les systèmes experts ne sont pas capables d'exercer un raisonnement créatif ou d'intuition. Leur capacité à résoudre des problèmes est strictement limitée par les règles qui leur ont été données. Par exemple, dans un domaine comme la médecine, un SE peut diagnostiquer une maladie en fonction de symptômes spécifiques, mais il ne pourra pas faire une hypothèse novatrice ou proposer un traitement expérimental.

#### 1.6. **Problèmes d'interprétabilité**
   - La logique de prise de décision dans un système expert peut parfois être difficile à comprendre pour les utilisateurs finaux. Même si un SE donne une réponse correcte, il peut être difficile d’expliquer pourquoi une certaine règle a été appliquée. Cela peut réduire la confiance dans le système, surtout dans des domaines sensibles comme la médecine ou le droit.

#### 1.7. **Problème de gestion des connaissances tacites**
   - Les systèmes experts sont principalement fondés sur des connaissances explicites (celles qui peuvent être formalisées sous forme de règles), mais ils échouent à capturer les connaissances tacites (les compétences et l’expérience de l'expert, qui ne sont pas facilement formulées). Cette limitation les rend particulièrement vulnérables lorsque les problèmes à traiter nécessitent une expérience approfondie ou une intuition que l'expert lui-même peut avoir développée.
   
   

### 1.8.Passage aux modèles d'apprentissage automatique

Les limites des systèmes experts, en particulier leur rigidité et leur dépendance aux règles explicites, ont conduit au développement des modèles d'apprentissage automatique (AA). Ces modèles, basés sur l'idée que les machines peuvent apprendre à partir de données et améliorer leurs performances au fil du temps, surmontent de nombreuses limitations des systèmes experts.

#### 1.8.1. **Adaptation et apprentissage à partir des données**
   - Contrairement aux systèmes experts, les modèles d'apprentissage automatique n'ont pas besoin de règles explicitement définies. Ils apprennent à partir de grandes quantités de données et peuvent ajuster leurs paramètres en fonction des patterns qu'ils découvrent dans les données. Cela leur permet de s'adapter plus facilement aux nouvelles situations et d’améliorer leur performance au fur et à mesure de l’introduction de nouvelles données.

#### 1.8.2. **Gestion de l'incertitude et des imprécisions**
   - Les modèles d'apprentissage automatique, et en particulier les réseaux neuronaux et les méthodes probabilistes, peuvent gérer l'incertitude et les données incomplètes. Par exemple, ils peuvent utiliser des techniques de régularisation pour prédire des résultats même en cas de données bruyantes ou manquantes.

#### 1.8.3. **Créativité et raisonnement**
   - Bien que les systèmes d'apprentissage automatique n'aient pas une "créativité" au sens humain du terme, ils peuvent découvrir des patterns complexes dans les données et proposer des solutions innovantes qui ne sont pas évidentes à partir des règles explicites. Par exemple, un modèle d'IA peut découvrir une nouvelle manière d’optimiser un processus industriel en analysant des données historiques.

#### 1.8.4. **Évolutivité et mise à jour automatique**
   - Les modèles d'apprentissage automatique peuvent être mis à jour facilement et automatiquement en traitant de nouvelles données. Cela permet à ces systèmes de rester pertinents au fur et à mesure que de nouvelles informations deviennent disponibles, sans nécessiter une intervention manuelle coûteuse.


Les systèmes experts sont utiles dans des contextes où les règles sont bien établies et les domaines d'application sont relativement stables. Cependant, leur rigidité, leur manque de capacité d'adaptation, et leur dépendance aux connaissances explicites font qu'ils sont dépassés par les capacités d'apprentissage et d'adaptation des modèles d'apprentissage automatique. Ces derniers offrent une plus grande flexibilité et un meilleur traitement de l'incertitude, tout en permettant un raisonnement plus créatif et autonome.



### 2. **Du Machine Learning**


Le machine learning, ou apprentissage automatique, est une branche de l’intelligence artificielle qui permet à une machine d’acquérir des connaissances par elle-même, sans nécessiter une programmation explicite. Cette approche a été introduite par Arthur Samuel en 1959, lorsqu’il a développé le premier programme de jeu de dames "intelligent". 

Ce programme avait la capacité d'apprendre à jouer sans que des stratégies prédéfinies d'attaque ou de défense ne soient intégrées dans le système. L'idée d’Arthur Samuel était de mettre deux machines en compétition, chacune jouant avec des stratégies initialement aléatoires, puis de récompenser celle qui gagnait chaque partie avec un système de points. Au fil des parties, les programmes apprenaient à identifier les situations favorables et défavorables, découvrant ainsi les stratégies gagnantes. Ce fut un succès !

Aujourd’hui, le machine learning est au cœur de nombreuses technologies modernes telles que ChatGPT, les voitures autonomes et divers systèmes dits "intelligents". Il est désormais l’une des disciplines principales de l’intelligence artificielle.


#### 2.1.Comment apprendre ?

L’idée centrale du machine learning est d’apprendre à partir de données. Pour cela, il existe trois grandes méthodes d’apprentissage :

#### 2.1.1. L'apprentissage supervisé

L'apprentissage supervisé est une méthode d'entraînement des machines où elles apprennent à partir d'exemples déjà connus. On lui fournit un ensemble de données où chaque élément est accompagné de la réponse correcte, ou "étiquette". Le but est que la machine découvre une règle générale qui lui permettra de prédire les réponses pour de nouvelles données qu'elle n'a jamais vues.

##### **Illustration simple**

Imaginons que vous vouliez apprendre à un programme informatique à reconnaître des fruits (des pommes). Vous lui montrez plusieurs images de fruits, comme des pommes, en lui disant quelle image correspond à quel fruit. Par exemple :

- **Image 1** : une pomme (étiquette "pomme")

![Texte alternatif](AS.png)

Le programme va apprendre à partir de ces exemples et va identifier des caractéristiques communes à chaque fruit, comme la couleur ou la forme. Une fois qu'il a vu assez d'exemples, il pourra reconnaître de nouvelles images qu’il n’a jamais vues auparavant, comme une autre pomme.

##### **Applications courantes**

- **Reconnaissance d'images** : Un programme peut apprendre à identifier des objets dans des photos, comme des voitures, des animaux ou des bâtiments.

- **Prédiction des prix** : En utilisant des données historiques, un programme peut apprendre à prédire le prix futur d’un produit en fonction de certaines caractéristiques, comme la taille, la couleur, ou la demande.


L'apprentissage supervisé est une méthode puissante qui repose sur l'exemple et l'expérience pour aider les machines à faire des prédictions intelligentes. C’est comme un apprentissage par imitation, où la machine se base sur des données étiquetées pour trouver des patterns et généraliser à de nouveaux cas.


La Machine étudie les exemples qu’on lui montre, et grâce à un système d’auto-évaluation et
d’auto-amélioration, elle parvient à apprendre à réaliser la tâche
qu’on lui demande d’effectuer.

Selon la nature de cette tâche, on distingue deux types de problèmes :

![Texte alternatif](RvsC.png)

| **Critères**                    | **Classification**                             | **Régression**                                  |
|-----------------------------|-----------------------------------------------|------------------------------------------------|
| **Objectif principal**       | Prédire une catégorie ou une étiquette parmi un ensemble de classes (discrètes). | Prédire une valeur continue (numérique).       |
| **Exemple de sortie**       | Étiquettes comme "chat", "chien", ou "voiture". | Valeurs comme le prix d'une maison ou la température. |

- **Classification** est utilisée lorsque vous avez des données qui appartiennent à des catégories distinctes.
- **Régression** est utilisée lorsque vous avez des données continues, et l’objectif est de prédire une valeur numérique précise.

#### 2.1.2. L'apprentissage non supervisé

L'apprentissage non supervisé est une méthode où la machine apprend à partir de données qui ne sont pas étiquetées, c'est-à-dire sans savoir à l'avance quelle est la réponse correcte. Au lieu de se baser sur des exemples avec des réponses connues, l'objectif est de découvrir des structures ou des patterns cachés dans les données.

**Illustration simple**
Imaginons que vous ayez une boîte pleine de fruits, mais cette fois, vous ne savez pas à l'avance lesquels sont des pommes, des oranges ou des bananes. Vous voulez que le programme trouve des groupes ou des catégories de fruits similaires sans lui dire ce qu'il doit chercher. Le programme pourrait, par exemple, regrouper les fruits par couleur ou par taille :

![Texte alternatif](AN.png)

Un groupe de fruits rouges (qui pourraient être des pommes et des cerises).
Un groupe de fruits oranges (qui pourraient être des oranges et des pêches).
Il n’a pas besoin de savoir qu’une pomme est une pomme ou qu’une orange est une orange. Il découvre des similarités tout seul et crée des groupes d’objets qui se ressemblent.

**Applications courantes**
- **Segmentation de clients** : Par exemple, dans le commerce, on peut utiliser l’apprentissage non supervisé pour regrouper les clients en différentes catégories en fonction de leurs achats, sans savoir à l’avance combien de groupes il y a ou qui appartient à chaque groupe.

- **Réduction de dimensions** : Dans des données complexes, l'apprentissage non supervisé peut être utilisé pour simplifier ces données tout en conservant les informations importantes. C'est comme si vous résumiez un livre en ne gardant que les idées principales.

L'apprentissage non supervisé est idéal lorsqu’on n’a pas de réponses précises à fournir à la machine. Il permet à la machine de trouver des modèles ou des relations cachées dans les données, de manière autonome. C'est comme laisser l’ordinateur explorer et découvrir des choses par lui-même.

Ce type d’application porte de le nom de **clustering**. Il s’agit de l’une des nombreuses applications de l’apprentissage non supervisé et démontre parfaitement son utilité. En effet, dans la pratique, il peut être très intéressant de laisser la machine proposer sa propre manière de faire les choses, plutôt que de la contraindre, de manière supervisée, à reproduire notre façon de faire.

#### 2.1.3. L'apprentissage par renforcement

L'apprentissage par renforcement est une méthode d'apprentissage où une machine apprend à prendre des décisions en interagissant avec un environnement. Contrairement à l'apprentissage supervisé ou non supervisé, il ne repose ni sur des données étiquetées ni sur des patterns cachés à découvrir. La machine apprend en recevant des **récompenses** ou des **punitions** en fonction des actions qu’elle effectue.

##### **Illustration simple**

Imaginez un robot qui doit apprendre à naviguer dans un labyrinthe pour atteindre la sortie. Voici comment fonctionne l'apprentissage par renforcement :

1. **Le robot agit** : Il décide de se déplacer à gauche, à droite, en avant ou en arrière.
2. **L'environnement réagit** : S'il se rapproche de la sortie, il reçoit une **récompense** (par exemple, +10 points). S'il frappe un mur ou s’éloigne de la sortie, il reçoit une **punition** (par exemple, -5 points).
3. **Apprentissage** : À force d’explorations, le robot découvre quelles actions lui rapportent des points positifs et apprend à choisir les mouvements qui maximisent sa récompense.

![Texte alternatif](AR.png)

##### **Composants principaux**

- **Agent** : La machine ou le programme qui prend les décisions (le robot dans l'exemple).
- **Environnement** : Le monde dans lequel l'agent agit (le labyrinthe).
- **Action** : Une décision prise par l'agent (avancer, reculer, tourner).
- **Récompense** : Un retour positif ou négatif que l’agent reçoit en fonction de ses actions.
- **État** : La situation actuelle de l'agent dans l'environnement.

##### **Applications courantes**

- **Jeux** : Les algorithmes d'apprentissage par renforcement sont célèbres pour avoir battu des champions humains aux échecs, au jeu de Go et même à des jeux vidéo complexes comme Dota 2.
- **Robots** : Enseigner à des robots à marcher, à manipuler des objets ou à naviguer dans des environnements dynamiques.
- **Voitures autonomes** : Aider une voiture à prendre des décisions en temps réel pour éviter des obstacles et atteindre sa destination.
- **Finance** : Optimisation des portefeuilles d’investissement en fonction de récompenses basées sur la rentabilité.

##### **Avantages**

- Permet d'apprendre dans des environnements complexes où il n'existe pas de solution explicite.
- Capable de s’adapter à des situations nouvelles et imprévues.

##### **Limites**

- **Temps d'apprentissage** : Peut nécessiter beaucoup de simulations ou d’essais pour arriver à un bon résultat.
- **Exploration vs Exploitation** : L'agent doit trouver un équilibre entre tester de nouvelles actions (exploration) et utiliser ce qu'il connaît déjà pour maximiser les récompenses (exploitation).
- **Risque d'apprentissage erroné** : Si les récompenses ne sont pas bien définies, l’agent peut adopter des comportements imprévus ou non désirés.

L'apprentissage par renforcement est une méthode puissante qui repose sur le principe d’apprentissage par essais et erreurs. En maximisant les récompenses, les agents deviennent capables de résoudre des problèmes complexes et de prendre des décisions optimales, souvent dans des situations où les autres types d’apprentissage sont moins efficaces.

**Pour résumé**  

Le machine learning se distingue de la programmation classique, où les développeurs doivent définir manuellement le fonctionnement des modèles. Cette capacité à apprendre directement à partir des données confère au machine learning un avantage notable sur les approches traditionnelles, notamment dans des domaines comme la vision par ordinateur ou les chatbots.  

Créer un modèle capable de reconnaître un objet sur une image, par exemple, serait extrêmement complexe pour un programmeur. Il faudrait concevoir un grand nombre de modèles géométriques et trigonométriques pour identifier des formes simples (droites, cercles, triangles) et les combiner afin de déterminer à quoi elles correspondent, tout en tenant compte de facteurs tels que la lumière, les ombres ou la perspective. Une tâche titanesque.  

En revanche, le machine learning simplifie ce processus en générant automatiquement un modèle optimal à partir des données fournies. Son objectif principal est de "bien s’adapter" aux données disponibles, ce qui explique son efficacité remarquable dans de telles applications.

### 3. Régression

Dans ce chapitre, vous découvrirez comment construire des modèles de régression. Comme expliqué dans le premier chapitre, ces modèles sont conçus pour prédire une variable quantitative \\( y \\) à partir de données d’entrée \\( X \\).  

Voici quelques exemples d’utilisation :  
- Estimer le prix d’une maison.  
- Calculer le temps de trajet d’un taxi.  
- Prédire le nombre de visiteurs quotidiens sur un site web.  
- Anticiper le chiffre d’affaires d’une entreprise pour l’année à venir.  

Ces modèles fonctionnent en mode supervisé, ce qui signifie qu’ils apprennent à partir d’exemples où les données d’entrée \\( X \\) sont associées à des valeurs cibles \\( y \\). L’objectif est de permettre à la machine de trouver une relation entre \\( X \\) et \\( y \\), pour effectuer des prédictions sur de nouvelles données.

![Texte alternatif](1.png)


**Mise en situation** 

Supposons que vous souhaitiez créer un modèle capable de prédire le prix d’une maison en fonction de sa surface habitable. 

![Texte alternatif](2.png)

Pour cela, vous disposez d’une base de données comprenant plusieurs exemples de maisons, avec pour chacune un prix \( y \) et une surface habitable \( X \). 

En représentant ces données sur un graphique, vous obtenez une visualisation de leur relation.  

![Texte alternatif](3.png)

Comme évoqué dans le point qui a précédé, l’objectif principal de l’apprentissage supervisé est de concevoir un modèle qui s’ajuste au mieux à ces données. Ce modèle pourra ensuite être utilisé pour faire des prédictions sur de nouvelles maisons.  

![Texte alternatif](4.png)

En machine learning, il existe de nombreux modèles pour accomplir cette tâche. Dans ce chapitre, nous allons explorer le plus simple d’entre eux : **la régression linéaire**.  


#### **Le modèle de régression linéaire**  
Le modèle de régression linéaire repose sur la formule bien connue :  
\\[ f(x) = wx + b \\]  

![Texte alternatif](5.png)

Lorsqu’on trace cette fonction, elle se présente sous la forme d’une droite. Dans cette formule :  
- \\( w \\) contrôle l’inclinaison de la droite (sa pente).  
- \\( b \\) correspond à l’ordonnée à l’origine (le point où la droite coupe l’axe \( y \)).  

![Texte alternatif](6.png)

L’objectif est de trouver les valeurs optimales de \\( w \\) et \\( b \\) pour que le modèle s’ajuste le mieux possible aux données \\( (X, y) \\).  

![Texte alternatif](7.png)

Dans une approche classique, c’est le rôle du développeur de déterminer ces paramètres. Cependant, en machine learning, cette tâche est automatisée. La machine utilise un algorithme d’optimisation pour ajuster \\( w \\) et \\( b \\), en minimisant les erreurs entre le modèle et les données observées.  

Pour commencer, il est essentiel de mesurer ces erreurs, souvent représentées graphiquement par des segments rouges reliant chaque point au modèle. Cette mesure est effectuée à l’aide d’une **fonction coût**, qui sert à quantifier la différence entre les prédictions du modèle et les valeurs réelles des données.  

En ajustant les paramètres pour minimiser cette fonction coût, la machine parvient à trouver la meilleure droite possible pour représenter les données.  

![Texte alternatif](8.png)


**La fonction coût**  

Calculer la fonction coût est une démarche simple qui consiste à évaluer les erreurs commises par le modèle pour chaque point de données, puis à calculer leur moyenne.  

1. **Calcul des résidus**  

Pour chaque point \\( i \\), on mesure l’écart entre la prédiction du modèle \\( f(x(i)) \\) et la valeur réelle \\( y(i) \\). Cet écart, appelé résidu \\( R(i) \\), se calcule ainsi :  
   \\[
   R(i) = f(x(i)) - y(i)
   \\]  
   Par exemple, si pour la première maison \\( x(1) = 95 \\) (surface habitable) et \\( y(1) = 130 \\) (prix), le modèle prédit \\( f(x(1)) = 150 \\), alors le résidu est :  
   \\[
   R(1) = 150 - 130 = 20
   \\]  
   Cela signifie que le modèle a fait une erreur de 20 000 euros.  

![Texte alternatif](9.png)

2. **Prévenir l’annulation des erreurs opposées**  

Lorsque des résidus positifs et négatifs sont additionnés, ils risquent de s’annuler, ce qui fausserait le calcul de l’erreur totale. Pour éviter cela, on élève chaque résidu au carré, ce qui garantit des valeurs toujours positives. Cela donne la **distance euclidienne** pour chaque point \\( D(i) \\) :  
   \\[
   D(i) = \left(f(x(i)) - y(i)\right)^2
   \\]  

![Texte alternatif](10.png)

3. **Calcul de la perte globale**  

Une fois les distances au carré calculées pour tous les points, on en fait la somme, puis on calcule la moyenne pour obtenir la perte globale du modèle (appelée **Loss** en anglais) :  
   \\[
   \text{Loss} = \frac{1}{2m} \sum_{i=1}^{m} \left(y(i) - f(x(i))\right)^2
   \\]  
   où \\( m \\) représente le nombre total de points dans le jeu de données.  

C’est ainsi que la fonction coût est obtenue, permettant de mesurer la performance du modèle et d’ajuster ses paramètres pour réduire les erreurs.

4. **Allure de la fonction coût**

Lorsqu’on fait varier les paramètres \\( w \\) et \\(b\\) de notre modèle, on constate que la fonction coût prend la forme d’une parabole. Le graphique ci-dessous illustre ce qui se produit lorsqu’on calcule la fonction coût pour différents modèles en faisant varier la valeur de \\(b\\).

![Texte alternatif](11.png)
![Texte alternatif](12.png)

Dans cet exemple, on voit clairement que le minimum de la fonction coût est atteint lorsque \\(b\\) avoisine la valeur \\(37\\).

![Texte alternatif](13.png)


La même chose se produit lorsqu’on fait varier l’inclinaison de notre modèle, c’est-à-dire le paramètre \\(w\\). Ici, on observe que le modèle fait les plus petites erreurs lorsque \\(w = 2:4\\)

![Texte alternatif](14.png)

Il est possible de visualiser ces deux graphiques en les combinant dans un graphique 3D. On
obtient alors une surface en forme de bol, au fond de laquelle se trouve le minimum de la
fonction coût, ainsi que les coordonnées \\((w; b)\\) permettant d’y accéder.

![Texte alternatif](15.png)

À première vue, il semble donc simple de trouver ce minimum. Il suffit de regarder où il se situe. Mais cela n’est possible que parce que nous avons créé cette carte en testant de nombreuses valeurs de \\((w; b)\\). Notre machine quant à elle, ignore complètement où se situe ce minimum, car elle ne dispose pas de cette carte. C’est pourquoi nous allons développer un algorithme qui lui permettra de trouver ce minimum automatiquement. Cet algorithme porte le nom de **descente de gradient**



**L’algorithme de descente de gradient**  

La descente de gradient est l’un des algorithmes les plus répandus en machine learning pour optimiser les modèles. Son principe repose sur le calcul du gradient de la fonction coût, c’est-à-dire la manière dont cette fonction évolue en réponse à de légères variations des paramètres \\( w \\) et \\( b \\). Ensuite, on ajuste ces paramètres en avançant dans la direction qui réduit la fonction coût, d’où le terme "descente de gradient".  

#### Fonctionnement  

L’algorithme débute par des valeurs initiales \\( (w_0, b_0) \\) choisies aléatoirement. Cela génère un premier modèle. À partir de ce point, on calcule le gradient, c’est-à-dire la direction de la pente de la fonction coût. 

![Texte alternatif](16.png)

Ensuite :  

1. **Premier ajustement :** On fait un pas dans la direction descendante, ce qui donne de nouveaux paramètres \\( (w_1, b_1) \\). Avec ces valeurs mises à jour, un nouveau modèle est créé, qui s’ajuste un peu mieux aux données \\( (X, y) \\).  

2. **Répétition :** À ce nouveau point, on recalcule le gradient et on ajuste à nouveau les paramètres.  

![Texte alternatif](17.png)

3. **Convergence :** En répétant ce processus, on finit par atteindre le minimum de la fonction coût, ce qui correspond au modèle optimal.  

![Texte alternatif](18.png)

Pour mieux comprendre, on peut comparer cela à un randonneur dans une montagne cherchant à atteindre la vallée. Sans carte, il suit simplement la pente descendante, jusqu’à atteindre le point le plus bas. Cette méthode fonctionne bien si la vallée n’a qu’un seul creux, comme c’est le cas pour une fonction coût bien définie.  

![Texte alternatif](19.png)

#### Formules mathématiques  
L’algorithme se décrit ainsi :  
\\[
w_{t+1} = w_t - \eta \frac{\partial L}{\partial w_t}  
\\]  
\\[
b_{t+1} = b_t - \eta \frac{\partial L}{\partial b_t}
\\]  
- \\( \frac{\partial L}{\partial w_t} \\) et \\( \frac{\partial L}{\partial b_t} \\) représentent les gradients de la fonction coût \\( L \\) par rapport aux paramètres \\( w \\) et \\( b \\).  
- \\( \eta \\) est la vitesse d’apprentissage, qui contrôle la taille des pas effectués à chaque itération.  
- \\( t \\) indique l’itération actuelle.  

### Exemple d’application  
Prenons \\( b \\) comme paramètre : 

##### Initialisation de l'algorithme d’apprentissage

Nous débutons l’algorithme d’apprentissage avec une valeur initiale de \\( b_0 \\) choisie au hasard. Ici, \\( b_0 = -102 \\).

![Texte alternatif](20.png)

##### Calcul du gradient

En calculant le gradient \\( \frac{\partial L}{\partial b} \\), on obtient une valeur négative (ce qui est logique, car la fonction diminue lorsque l'on augmente la valeur de \\( b \\)). 

Pour simplifier l'explication et éviter des calculs inutiles, supposons que :  
\\( \frac{\partial L}{\partial b} = -2 \\).

![Texte alternatif](21.png)

##### Formule de la descente de gradient

Ainsi, la formule de la descente de gradient donne :  
\\[ b_1 = b_0 - \eta \frac{\partial L}{\partial b_0} \\]  
\\[ b_1 = b_0 - 1 \times (-2) \\]  
\\[ b_1 = b_0 + 2 \\]  

Ainsi, \\( b_1 \\) est donc supérieur à \\( b_0 \\).

![Texte alternatif](22.png)

##### Répétition de l'algorithme

On répète ainsi l’algorithme en recalculant le gradient et en mettant à jour les paramètres \\( w \\) et \\( b \\) :  
\\[ w_{t+1} = w_t - \eta \frac{\partial L}{\partial w_t} \\]  
\\[ b_{t+1} = b_t - \eta \frac{\partial L}{\partial b_t} \\]  

![Texte alternatif](23.png)

...jusqu’à ce que la convergence soit atteinte, c’est-à-dire lorsque les changements de la fonction coût deviennent insignifiants.  

![Texte alternatif](24.png) 

### Importance et applications  
La descente de gradient est bien plus qu’un outil pour la régression linéaire. Elle est au cœur de nombreux algorithmes d’optimisation utilisés en deep learning, notamment pour entraîner les réseaux de neurones.  


### Implémentation Python


Pour développer un modèle de régression linéaire avec Python, nous allons utiliser le package scikit-learn, très populaire pour faire du machine learning.

![Texte alternatif](25.png) 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

##### Simulation des données avec NumPy

Pour commencer, simulons des données à l'aide de la librairie NumPy. Nous allons générer 50 points qui suivent la relation :  

\\[ y = 1.5x + 3 \\]

Ces données serviront de base pour notre modèle.


```python
# Fixer la graine aléatoire pour garantir la reproductibilité des résultats
np.random.seed(0)

# Définir le nombre de points de données
m = 50

# Générer 50 points aléatoires suivant une distribution normale, puis les redimensionner en un tableau colonne
X = np.random.randn(m).reshape(-1, 1)

# Générer les valeurs correspondantes de y en utilisant la relation y = 1.5x + 3, avec un bruit ajouté
y = 1.5 * X + 3 + np.random.randn(m).reshape(-1, 1)

# Visualiser les données avec un nuage de points
plt.scatter(X, y)  # Tracer les points (X, y) sur un graphique
plt.show()  # Afficher le graphique
```

1. **`np.random.seed(0)`** : Fixe la graine aléatoire pour s'assurer que les résultats soient reproductibles à chaque exécution du code.
2. **`m = 50`** : Définit le nombre de points de données à générer (50 dans ce cas).
3. **`X = np.random.randn(m).reshape(-1, 1)`** : Génère un tableau de 50 valeurs suivant une distribution normale standard (moyenne = 0, écart-type = 1), puis les réorganise en une colonne.
4. **`y = 1.5 * X + 3 + np.random.randn(m).reshape(-1, 1)`** : Calcule les valeurs correspondantes de `y` selon l'équation \( y = 1.5x + 3 \), avec l'ajout d'un bruit gaussien pour simuler des données réalistes.
5. **`plt.scatter(X, y)`** : Crée un nuage de points pour visualiser les paires de données \( (X, y) \).
6. **`plt.show()`** : Affiche le graphique pour que les données soient visibles.

Pour créer notre modèle de machine learning, nous allons utiliser le modèle `LinearRegression`  
de scikit-learn. Ce modèle n’utilise pas la descente de gradient vue dans ce chapitre, mais une  
autre méthode d’optimisation, plus rapide, que je détaille sur ma chaîne YouTube Machine  
Learnia.  

Une fois le modèle créé, nous utilisons la méthode `fit()` pour entraîner le modèle sur les données  
(X; y).  

Une fois le modèle entraîné, il suffit de l’utiliser pour faire des prédictions à l’aide de la méthode  
`predict()`.  


```python
# Création et entraînement du modèle de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Génération de nouvelles données pour la prédiction
new_data = np.linspace(-3, 3, 100).reshape(-1, 1)
predictions = model.predict(new_data)

# Visualisation des données et du modèle entraîné
plt.scatter(X, y, label="Données d'origine")
plt.plot(new_data, predictions, color="red", label="Modèle prédit")
plt.legend()
plt.show()
```

1. **Importation des bibliothèques nécessaires** : 
   - `LinearRegression` pour le modèle.
   - `numpy` pour manipuler les données.
   - `matplotlib.pyplot` pour la visualisation.

2. **Création et entraînement du modèle** :
   - La ligne `model = LinearRegression()` initialise un modèle de régression linéaire.
   - La méthode `fit(X, y)` entraîne ce modèle sur les données `X` et `y`.

3. **Création de nouvelles données pour prédiction** :
   - `np.linspace(-3, 3, 100)` génère 100 points régulièrement espacés entre -3 et 3.
   - `.reshape(-1, 1)` transforme ces points en un tableau de dimension `(100, 1)` compatible avec le modèle.

4. **Prédiction avec le modèle entraîné** :
   - La méthode `predict(new_data)` retourne les prédictions pour les nouvelles données `new_data`.

5. **Visualisation** :
   - `plt.scatter(X, y)` affiche les données originales sous forme de points.
   - `plt.plot(new_data, predictions)` trace la ligne représentant les prédictions du modèle.
   - `plt.legend()` ajoute une légende pour distinguer les éléments affichés.
   - `plt.show()` affiche le graphique.

#### Modèle de Nearest Neighbors


L’algorithme du Nearest Neighbors est particulier, car il nécessite de garder en mémoire tous les points du jeu de données pour calculer la distance entre ces points et un nouveau point dont on souhaite faire une prédiction.

Par exemple, voici un point dont on veut connaître la classe.

![Texte alternatif](30.png) 

Pour cela, on calcule la distance entre ce point
et tous les autres points du jeu de données. Il
existe plusieurs méthodes pour calculer la distance
entre deux points, mais bien souvent, on
utilise la distance euclidienne (rappelez-vous,
c’est la distance que nous avons vue dans le
chapitre sur la régression)

![Texte alternatif](31.png) 

En classant chaque distance de la plus petite à la
plus grande, on identifie le point du jeu de données
le plus proche. Dès lors, on attribue notre
point à la même classe que son plus proche voisin.

![Texte alternatif](32.png) 

Nous pouvons généraliser cette approche en
appliquant cette opération à tous les points de
notre espace...

![Texte alternatif](33.png) 

... Ce qui permet de mettre en avant la frontière
de décision, c’est-à-dire la frontière qui sépare
les deux classes.

![Texte alternatif](34.png) 

Et voilà, vous savez désormais comment fonctionne l’algorithme du plus proche voisin. Malgré sa simplicité, cet algorithme est encore largement utilisé aujourd’hui dans de nombreuses applications, y compris chez Spotify, Netflix, Amazon, et bien d’autres.



##### Généralisation : Le modèle du K-Nearest Neighbors


Dans sa forme la plus simple, l’algorithme du plus proche voisin, dans sa version à 1 voisin,
n’est pas très robuste. En effet, voyons ce qu’il se produit lorsqu’on fait face à un jeu de données
un peu plus complexe. Ci-dessous, quelques valeurs aberrantes se sont glissées dans chaque
classes. Cependant, à vue d’oeil, cela ne change pas vraiment la tendance globale qui suggère
que les emails spams sont situés en haut à droite, et les non-spams en bas à gauche.

![Texte alternatif](35.png) 

Malheureusement, la frontière de décision obtenue en suivant l’approche du plus proche voisin
donne un résultat chaotique, qui ne correspond plus du tout à la tendance globale que nous
avions observée précédemment. L’email de départ dont nous voulions connaître la classe est
maintenant classé parmi les emails non-spams bien qu’il semble plus proche du groupe des
spams.

![Texte alternatif](36.png) 

Pour éviter ce problème, on utilise généralement les k-voisins plus proches, pour ensuite
assimiler notre point à la classe majoritaire parmi ces k-voisins. 

Voici un exemple avec k=3. Ici, notre point compte parmi ses 3 voisins les plus proches, 2 points
de la classe Orange et 1 point de la classe Gris foncé. Il est donc assimilé à la classe Orange. La
frontière de décision est ainsi plus cohérente que précédemment.

![Texte alternatif](37.png) 

Et voici un exemple avec K=5.

![Texte alternatif](38.png) 

Comme on peut le constater, plus le nombre de voisins augmente, moins la frontière de décision est sensible aux valeurs aberrantes.

Cependant, il faut rester vigilant à ne pas trop augmenter le nombre de voisins, au risque de
trop lisser la frontière de décision. L’exemple ci-dessous montre un jeu de données dans lequel
les deux classes prennent une forme de croissant, avec une légère superposition de certains
points.

![Texte alternatif](39.png) 

Ainsi, la situation où \( k = 1 \) n’est pas idéale, car elle est trop sensible aux valeurs aberrantes, comme nous l’avons vu précédemment. Mais une valeur de \( k \) trop élevée a pour effet de tracer une frontière de décision qui ne correspond plus du tout à l’allure générale de notre jeu de données.  

Il convient donc de trouver le juste milieu.

#### Implémentation avec Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

1. **`import numpy as np`**  
   - Importe la bibliothèque NumPy, utilisée pour les opérations mathématiques et la manipulation de tableaux.  
   - L'alias `np` est une convention courante pour simplifier son utilisation dans le code.  

2. **`import matplotlib.pyplot as plt`**  
   - Importe la bibliothèque Matplotlib, qui est utilisée pour créer des graphiques et visualiser les données.  
   - L'alias `plt` est également une convention pour faciliter son utilisation.  

3. **`from sklearn.datasets import make_blobs`**  
   - Importe la fonction `make_blobs` de Scikit-learn, une bibliothèque populaire de machine learning.  
   - `make_blobs` est utilisée pour générer artificiellement des données de type "cluster", souvent utilisées pour tester ou illustrer des algorithmes de classification.  

4. **`from sklearn.neighbors import KNeighborsClassifier`**  
   - Importe la classe `KNeighborsClassifier`, qui implémente l'algorithme des \( k \)-plus proches voisins (k-Nearest Neighbors).  
   - Cette classe permet de créer un modèle de classification basé sur les distances entre les points dans un espace de caractéristiques.  

##### Simuler des données

Commençons par simuler des données. Nous allons :  
- Créer deux classes \\( y \\), avec 50 points.  
- Générer deux variables \\( X \\).  
- Ajouter des points dont la classe est inconnue.  

```python
# Générer des données pour deux classes (X, y)
X, y = make_blobs(
    n_samples=50,          # Nombre total de points
    centers=2,             # Nombre de classes (ou centres)
    n_features=2,          # Nombre de variables/features
    random_state=0,        # Pour assurer la reproductibilité
    cluster_std=1          # Écart type des clusters
)

# Générer 10 nouveaux points avec des classes inconnues
X_new = np.random.uniform(
    low=X.min(),           # Valeur minimale des données existantes
    high=X.max(),          # Valeur maximale des données existantes
    size=(10, 2)           # 10 points, chacun ayant deux dimensions
)

# Affichage des points existants
plt.scatter(
    X[:, 0],               # Première dimension des données existantes
    X[:, 1],               # Deuxième dimension des données existantes
    c=y,                   # Couleur en fonction des classes
    cmap="bwr"             # Palette de couleurs bleu-rouge
)

# Affichage des nouveaux points avec une apparence différente
plt.scatter(
    X_new[:, 0],           # Première dimension des nouveaux points
    X_new[:, 1],           # Deuxième dimension des nouveaux points
    c="white",             # Couleur blanche
    edgecolor="k"          # Contour noir
)

# Afficher le graphique
plt.show()
```

1. **Génération des données avec `make_blobs` :**
   - On crée 50 points appartenant à deux classes distinctes (définies par les centres des clusters).
   - Les données ont deux variables (\(X_1\) et \(X_2\)) avec une certaine dispersion contrôlée par `cluster_std`.

2. **Génération des nouveaux points avec `np.random.uniform` :**
   - Ces points sont créés aléatoirement dans l'intervalle de valeurs de \( X \).
   - Ils n'ont pas de classe associée.

3. **Affichage des données existantes :**
   - Chaque point est coloré en fonction de sa classe grâce au paramètre `c=y` et la palette de couleurs `cmap="bwr"`.

4. **Affichage des nouveaux points :**
   - Ils apparaissent en blanc avec un contour noir pour les différencier des données existantes.

5. **Affichage du graphique :**
   - Les deux ensembles de points sont superposés sur le même plan.

Pour entraîner le modèle de `KNeighborsClassifier`, nous allons l’initialiser avec une valeur \\( k = 3 \\),  
puis utiliser la méthode `fit()` comme vu dans le chapitre sur la régression.


```python
model = KNeighborsClassifier(n_neighbors=3)  
model.fit(X, y)
```

Enfin, pour effectuer des prédictions et classer les nouveaux points de notre jeu de données, on utilise la méthode `predict()`. On arrive ainsi à classer les nouveaux points du jeu de données.

```python
predictions = model.predict(X_new)

# Utilisation d'une colormap plus visible
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='plasma')  # Utilisation de 'plasma' pour une meilleure visibilité
plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, cmap='plasma', edgecolor='k')
plt.show()
```

1. **`predictions = model.predict(X_new)`** :
   - Cette ligne utilise le modèle de classification (dans ce cas, un `KNeighborsClassifier`) pour prédire les classes des nouveaux points `X_new`. La méthode `predict()` applique le modèle à ces points et retourne les classes prédites.

2. **`plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4)`** :
   - Cette ligne crée un nuage de points (scatter plot) pour les données d'origine `X`. Les coordonnées des points sont données par `X[:, 0]` (pour les abscisses) et `X[:, 1]` (pour les ordonnées).
   - `c=y` colore chaque point en fonction de la classe réelle associée à chaque point, stockée dans `y`.
   - `alpha=0.4` définit la transparence des points, afin qu’ils soient plus translucides, ce qui permet de mieux voir la superposition des points dans le graphique.

3. **`plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, edgecolor='k')`** :
   - Cette ligne génère un second nuage de points pour afficher les nouveaux points `X_new`, dont les classes ont été prédites dans la ligne précédente. 
   - Les coordonnées sont extraites de `X_new[:, 0]` et `X_new[:, 1]` pour les abscisses et ordonnées, respectivement.
   - `c=predictions` colore les nouveaux points en fonction des classes qui ont été prédites pour ces points.
   - `edgecolor='k'` ajoute une bordure noire autour des points, afin de les distinguer visuellement.

4. **`plt.show()`** :
   - Cette ligne affiche le graphique complet avec les deux ensembles de points : les points d'origine (avec leurs classes réelles) et les nouveaux points (avec leurs classes prédites). La méthode `show()` est utilisée pour rendre le graphique visible.
